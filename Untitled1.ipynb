```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/diabetes.csv')
df
```

```python
df=df.drop(columns= ['patient_number'])
df
```

```python
df['chol_hdl_ratio'] = df['chol_hdl_ratio'].astype(str).str.replace(',', '.').astype(float)
df
```

```python
df = df.rename(columns={"hdl_chol": "hdl_cholesterol", "bmi": "BMI"})
df
```

```python
df['BMI'] = df['BMI'].str.replace(',', '.').astype(float)
def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underwt"
    elif 18.5 <= bmi < 24.9:
        return "Normal Wt"
    elif 24.9 <= bmi < 29.9:
        return "Overwt"
    else:
        return "Obese"
df['bmi_category'] = df['BMI'].apply(categorize_bmi)
print(df)
```

```python
df['chol_hdl_ratio'] = df['cholesterol'] / df['hdl_cholesterol']
print(df.head())
```

```python
df = df.drop_duplicates()
print(df)
```

```python
numerical_columns = ["cholesterol", "glucose", "hdl_cholesterol", "chol_hdl_ratio", "age", "height", "weight", "BMI", "systolic_bp", "diastolic_bp", "waist", "hip", "waist_hip_ratio"]
outliers = pd.DataFrame(columns=numerical_columns)
for column in numerical_columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        column_outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outliers = pd.concat([outliers, column_outliers])
print("Outliers in the dataset:")
print(outliers)
```
